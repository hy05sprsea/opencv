```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("imori.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.subplot(1, 2, 1)
plt.title("input")
plt.imshow(img)
plt.subplot(1, 2, 2)
plt.title("gray")
plt.imshow(img_gray, "gray")
plt.show()
```
