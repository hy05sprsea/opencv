```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread("imori.jpg")
img = cv2.resize(img, (258, 258))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

plt.imshow(img_gray, "gray")
plt.show()
```

```python
plt.hist(img_gray.ravel(), bins=255, rwidth=0.5, range=(0, 255))
plt.text(0, 500, "class0", fontsize=20)
plt.text(200, 500, "class1", fontsize=20)
plt.vlines(125, 0, 800, color="red", linestyles="dotted")
plt.xlabel = ("pixel")
plt.ylabel = ("appearance")
plt.show()
```

```python
th, img_th = cv2.threshold(img_gray, 0, 255, cv2.THRESH_OTSU)
print("threshold", th)
plt.imshow(img_th, 'gray')
```
