```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("imori_256x256.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_Gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(12, 3))
plt.subplot(1, 2, 1)
plt.imshow(img)
plt.show()
plt.figure(figsize=(12, 3))
plt.subplot(1, 2, 2)
plt.imshow(img_Gray, 'gray')
plt.show()
```
