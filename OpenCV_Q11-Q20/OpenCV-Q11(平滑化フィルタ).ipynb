```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img_noise = cv2.imread("imori_256x256_noise.png")
img_noise = cv2.cvtColor(img_noise, cv2.COLOR_BGR2RGB)
plt.imshow(img_noise)
plt.show()
```

```python
img_smoth = cv2.blur(img_noise, (5, 5))

plt.imshow(img_smoth)
plt.show()
```
