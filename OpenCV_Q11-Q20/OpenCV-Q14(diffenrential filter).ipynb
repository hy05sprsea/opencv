・微分フィルタ

高度の急激な変化が起こっている部分のエッジを取り出すフィルタ

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("imori_256x256.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()
```

```python
img_Gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.imshow(img_Gray, 'gray')
plt.show()
```

```python
kernel_x = np.array([[0, 0, 0],
                     [0, -1, 1],
                     [0, 0, 0]])
x = cv2.filter2D(img_Gray, -1, kernel_x)
plt.imshow(x, 'gray')  
```

```python
kernel_y = np.array([[0, -1, 0],
                     [0, 1, 0], 
                     [0, 0, 0]])
y = cv2.filter2D(img_Gray, -1, kernel_y)
plt.imshow(y, 'gray')      
```

```python
def filter_prewitt(img, k_size=(3, 3), mode='x'):
    k = np.zeros(k_size)
    if mode == 'x':
        k[:, 0] = 1
        k[:, -1] = -1
    else:
        k[0] = 1
        k[-1] = -1
    return cv2.filter2D(img, -1, k)

img_prewitt_x = filter_prewitt(img_Gray, mode='x')
img_prewitt_y = filter_prewitt(img_Gray, mode='y')

plt.subplot(1, 2, 1)
plt.imshow(img_prewitt_x, 'gray')
plt.subplot(1, 2, 2)
plt.imshow(img_prewitt_y, 'gray')
```
