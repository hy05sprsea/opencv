```python
from skimage import io
import matplotlib.pyplot as plt
import cv2
import numpy as np

img_noise_orig = io.imread('https://yoyoyo-yo.github.io/Gasyori100knock/assets/imori_256x256_noise.png')

plt.imshow(img_noise_orig)
plt.show()
```

```python
img_med = cv2.medianBlur(img_noise_orig, 3)
img_med2 = cv2.medianBlur(img_noise_orig, 5)

plt.subplot(1, 2, 1)
plt.imshow(img_med)

plt.subplot(1, 2, 2)
plt.imshow(img_med2)
plt.show()
```
